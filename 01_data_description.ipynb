# Data Description

This notebook loads both suicide rate datasets and computes descriptive statistics and data quality checks for the report.

We look at two datasets covering global suicide rates from 1990–2022, checking things like missing values, duplicates, and basic stats to understand what we're working with before cleaning.

```python
%pip install pandas
# pandas is the main library for loading and working with tabular data
import pandas as pd
```

---
## Dataset 1: suicide_rates_1990-2022.csv

### Load the dataset and check its shape

```python
# Load the granular dataset (broken down by country, year, sex, age group, generation)
df1 = pd.read_csv("suicide_rates_1990-2022.csv")

# Quick check: how big is the data?
print("Shape (rows, columns):", df1.shape)
df1.head()
```

### Column names and data types

```python
print(df1.dtypes)
```

---
### Descriptive Statistics — Numerical Columns

```python
# Pick out the numeric columns to get summary stats (mean, std, min, max, etc.)
num_cols1 = [
    "SuicideCount",
    "CauseSpecificDeathPercentage",
    "DeathRatePer100K",
    "Population",
    "GDP",
    "GDPPerCapita",
    "GrossNationalIncome",
    "GNIPerCapita",
    "InflationRate",
    "EmploymentPopulationRatio"
]

stats1 = df1[num_cols1].describe(percentiles=[0.5])
stats1 = stats1.rename(index={"50%": "median"})
stats1
```

### Descriptive Statistics — Categorical Columns

```python
# Loop through categorical columns to see unique values and most common entries
cat_cols1 = ["RegionCode", "RegionName", "CountryCode", "CountryName", "Sex", "AgeGroup", "Generation"]

for col in cat_cols1:
    print("Column:", col)
    print("Unique values:", df1[col].nunique())
    print("Top 3 most frequent:")
    print(df1[col].value_counts().head(3))
    print()
```

---
### Data Quality Checks

#### Missing Values

```python
# Check how many values are missing in each column
missing_count1 = df1.isnull().sum()
missing_pct1 = (df1.isnull().sum() / len(df1)) * 100

missing_table1 = pd.DataFrame({
    "Missing Count": missing_count1,
    "Missing %": missing_pct1.round(2)
})

# Only show columns that actually have missing values
print(missing_table1[missing_table1["Missing Count"] > 0])
```

#### Duplicate Rows

```python
# Check for exact duplicate rows in the dataset
num_duplicates1 = df1.duplicated().sum()
print("Number of duplicate rows:", num_duplicates1)
```

#### Negative / Illogical Values
Suicide counts, death rates, and population values cannot be negative.

```python
# Sanity check: these columns should never be negative
negative_suicide1 = df1[df1["SuicideCount"] < 0]
print("Rows with negative SuicideCount:", len(negative_suicide1))

negative_rate1 = df1[df1["DeathRatePer100K"] < 0]
print("Rows with negative DeathRatePer100K:", len(negative_rate1))

negative_pop1 = df1[df1["Population"] < 0]
print("Rows with negative Population:", len(negative_pop1))
```

#### Zero Suicide Counts

```python
# Count rows where suicide count is exactly zero (could be real or just missing data)
zero_suicides1 = df1[df1["SuicideCount"] == 0]
print("Rows with zero SuicideCount:", len(zero_suicides1))
zero_suicides1.head()
```
